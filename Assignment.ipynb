```python
# Step 1: data extract from web

url = 'https://insights.blackcoffer.com/ai-in-healthcare-to-improve-patient-outcomes/'

# Initialize newspaper3k 

article = newspaper.Article(url)
article.download()
article.parse()
```

```python
# Step 2: Text Preprocessing

text = article.text.lower()
text = text.replace('\n', ' ')
text = text.strip()

# Step 3: Sentence Tokenization

sentences = sent_tokenize(text)

# Step 4: Word Tokenization

words = [word for sent in sentences for word in word_tokenize(sent)]
```

```python
# Step 5: Variable Calculations
positive_words = ['a+','abound','abounds','abundance','abundant','accessable','accessible','acclaim','acclaimed',
                  'acclamation','accolade','accolades','accommodative','accomodative','accomplish','accomplished',
                  'accomplishment','accomplishments','accurate','accurately','achievable','achievement','achievements',
                  'achievible','acumen','adaptable','adaptive','adequate','adjustable','admirable','admirably','admiration',
                  'admire','admirer']

negative_words = ['2-faced', '2-faces', 'abnormal', 'abolish', 'abominable', 'abominably', 'abominate', 'abomination', 
                  'abort', 'aborted', 'aborts', 'abrade', 'abrasive', 'abrupt', 'abruptly', 'abscond', 'absence', 
                  'absent-minded', 'absentee', 'absurd', 'absurdity', 'absurdly', 'absurdness', 'abuse', 'abused', 
                  'abuses', 'abusive', 'abysmal', 'abysmally', 'abyss', 'accidental', 'accost', 'accursed', 'accusation', 
                  'accusations', 'accuse', 'accuses', 'accusing', 'accusingly', 'acerbate', 'acerbic', 'acerbically', 
                  'ache', 'ached', 'aches', 'achey', 'aching', 'acrid', 'acridly', 'acridness', 'acrimonious', 
                  'acrimoniously', 'acrimony']

complex_threshold = 3  # Define the word length threshold for complexity

positive_score = sum(word in positive_words for word in words)
negative_score = sum(word in negative_words for word in words)
polarity_score = (positive_score - negative_score) / len(words)
subjectivity_score = TextBlob(text).sentiment.subjectivity

average_sentence_length = len(words) / len(sentences)

stopwords_set = set(stopwords.words('english'))
complex_words = [word for word in words if len(word) > complex_threshold and word not in stopwords_set]
percentage_complex_words = (len(complex_words) / len(words)) * 100

fog_index = (average_sentence_length + percentage_complex_words) * 0.4

average_words_per_sentence = len(words) / len(sentences)

complex_word_count = len(complex_words)
word_count = len(words)
```

```python
# Helper function to count syllables in a word
def count_syllables(word):
    return max(1, len(nltk.tokenize.word_tokenize(word)))

syllables_per_word = sum(count_syllables(word) for word in words) / len(words)

personal_pronouns = ['i', 'you', 'he', 'she', 'we']  # Add more personal pronouns as needed
personal_pronoun_count = sum(word.lower() in personal_pronouns for word in words)

average_word_length = sum(len(word) for word in words) / len(words)
```
