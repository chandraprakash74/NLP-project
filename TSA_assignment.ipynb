### Lazy Regressor

```python
from sklearn.model_selection import train_test_split
#from lazypredict.Supervised import LazyRegressor

# Split the data into features (X) and target (y)
X = df[['WTISPLC','MCOILBRENTEU','GASREGM','IMPCH','EXPCH','PRUBBUSDM','WPUFD4111','PCU325211325211','PCU32611332611301']]
y = df['Domestic Market (Contract) Blow Molding, Low']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=3)
```

```python
from lazypredict.Supervised import LazyRegressor

# Define and fit the LazyRegressor model
model_lazy = LazyRegressor(verbose=0, ignore_warnings=True, custom_metric=None)
models,predicts = model_lazy.fit(X_train,X_test,y_train,y_test)
```

### Gradient Boosting Regressor

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.ensemble import GradientBoostingRegressor
```

```python
model_gb = GradientBoostingRegressor()
model_gb.fit(X_train, y_train)
```

```python
# Make predictions using the Gradient Boosting Regression model
y_pred_gb = model_gb.predict(X_test)
```

```python
# Calculate directional accuracy (DA)
da_gb = np.mean((y_pred_gb > y_test) == (y_pred_gb > y_test.shift()))
da_gb
```

```python
# Calculate RMSE (Root Mean Squared Error)
rmse_gb = np.sqrt(mean_squared_error(y_test, y_pred_gb))
rmse_gb
```

```python
# Calculate MAPE (Mean Absolute Percentage Error)
mape_gb = np.mean(np.abs((y_test - y_pred_gb) / y_test)) * 100
mape_gb
```

```python
# Calculate R2 score
r2_gb = model_gb.score(X_test, y_test)
r2_gb
```

```python
# Calculate MAE (Mean Absolute Error)
mae_gb = mean_absolute_error(y_test, y_pred_gb)
mae_gb
```

```python
# checking stationary
def checkStationarity(data):
    pvalue = adfuller(data)[1]
    if(pvalue>0.05):
        msg = 'p-value={}. Data is not stationary'.format(pvalue)
    else:
        msg='p-value={}. Data is stationary'.format(pvalue)
        
    return(msg)
```

```python
model1=ARIMA(df1['Domestic Market (Contract) Blow Molding, Low'],order = (2,1,1))
```

```python
r1 = model1.fit()
```

```python
r1.summary()
```

```python
# apply ljung box test

pvalue = sm.stats.acorr_ljungbox(r1.resid,lags=[1],return_df = True)['lb_pvalue'].values
print(pvalue)
if pvalue < 0.05:
    print('Reject H0. Bad Model')
else:
    print('Fail to reject H0. Hence the selected model is good model')
```
